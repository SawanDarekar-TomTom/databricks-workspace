```python
import json
from shapely.geometry import shape
from shapely.ops import unary_union

with open("/Users/sawan.darekar/Downloads/export.geojson") as f:
    gj = json.load(f)

geoms = [shape(feat["geometry"]) for feat in gj["features"]]
boundary = unary_union(geoms)          # merge into one (Multi)Polygon

wkt = boundary.wkt
with open("hungary_boundary.wkt", "w") as f:
    f.write(wkt)
```
