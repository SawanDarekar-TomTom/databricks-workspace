```python
%scala
import org.apache.spark.sql.SparkSession
import org.apache.spark.sql.functions._

val spark = SparkSession.builder.appName("DropDuplicates Example").getOrCreate()

// Sample dataset
val data = Seq(
  (1, "Alice", "2023-01-01", true),
  (2, "Bob", "2023-01-02", true),
  (1, "Alice", "2023-01-03", true), // Duplicate
  (3, "Charlie", "2023-01-03", true),
  (2, "Bob", "2023-01-04", true)    // Duplicate
)

val df = spark.createDataFrame(data).toDF("id", "name", "date")

// Show original DataFrame
println("Original DataFrame:")
df.show()

// Apply dropDuplicates on 'id' columns
val deduplicatedDF = df.dropDuplicates("id")

// Show DataFrame after dropping duplicates
println("DataFrame after dropping duplicates:")
deduplicatedDF.show()
```
