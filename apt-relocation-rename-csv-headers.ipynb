```python
# Cell 1: Read folder, load all CSVs, print row count per file
import os
import pandas as pd

FOLDER_PATH = "/Users/sawan.darekar/Desktop/workspace_data/SEACO-5792-USA-DC-Execution/BUILDING_FOOTPRINT_23_March"  # <-- update this path

csv_files = sorted([f for f in os.listdir(FOLDER_PATH) if f.endswith(".csv")])

dataframes = {}
original_counts = {}

print(f"Found {len(csv_files)} CSV file(s) in: {FOLDER_PATH}\n")
for fname in csv_files:
    fpath = os.path.join(FOLDER_PATH, fname)
    df = pd.read_csv(fpath)
    dataframes[fname] = df
    original_counts[fname] = len(df)
    print(f"  {fname}: {len(df)} rows, {len(df.columns)} columns")
```

```python
# Cell 2: Rename headers and overwrite CSV files
HEADER_RENAMES = {
    "Target_Value_old_xy": "Target_Value",
    "Orbis_X_New": "Orbis_X",
    "Orbis_Y_New": "Orbis_Y",
}

renamed_dataframes = {}

for fname, df in dataframes.items():
    original_cols = list(df.columns)
    df_renamed = df.rename(columns=HEADER_RENAMES)
    renamed_dataframes[fname] = df_renamed

    applied = {old: new for old, new in HEADER_RENAMES.items() if old in original_cols}
    skipped = {old: new for old, new in HEADER_RENAMES.items() if old not in original_cols}

    # Overwrite the file with renamed headers
    fpath = os.path.join(FOLDER_PATH, fname)
    df_renamed.to_csv(fpath, index=False)

    print(f"{fname}:")
    if applied:
        for old, new in applied.items():
            print(f"  Renamed: '{old}' -> '{new}'")
    else:
        print("  No matching headers found to rename")
    if skipped:
        for old, new in skipped.items():
            print(f"  Skipped (not found): '{old}'")
    print()

print("All files saved.")
```

```python
# Cell 3: Validate row counts are unchanged after rename
print("Row count validation:\n")
all_passed = True

for fname in csv_files:
    fpath = os.path.join(FOLDER_PATH, fname)
    df_after = pd.read_csv(fpath)
    after_count = len(df_after)
    before_count = original_counts[fname]

    status = "PASS" if after_count == before_count else "FAIL"
    if status == "FAIL":
        all_passed = False

    print(f"  {fname}: before={before_count}, after={after_count} -> {status}")

print()
if all_passed:
    print("All files passed: row counts unchanged.")
else:
    print("WARNING: Some files have row count mismatches!")
```
