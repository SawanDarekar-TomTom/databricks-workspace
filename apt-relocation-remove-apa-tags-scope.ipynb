# APT Relocation — Remove APA Tags Scope

This notebook scopes APTs that need their APA (Address Point Accuracy) tags removed as part of the relocation process.

## What this notebook does

1. **Install dependencies** — installs `pandas` via `requirements.txt`
2. **Read `relocation.csv`** — loads the list of relocated APTs using plain Python (pandas), each row contains a `Product_orbis_id` and relocation metadata
3. **Generate `relocation-scope.csv`** — for every `Product_orbis_id` in the input, creates 4 rows targeting the APA tags to be removed:
   - `metadata:apa:accuracy`
   - `metadata:apa:previous_location`
   - `metadata:apa:source`
   - `metadata:apa:updated`

## Input / Output

| File | Description |
|------|-------------|
| `relocation.csv` | Source file with relocated APT IDs and location metadata |
| `relocation-scope.csv` | Output scoping file consumed by the data correctness pipeline |

### Load Relocation CSV

```python
import pandas as pd

# csv_path = "/Users/sawan.darekar/Desktop/tomtom_git_workspace/databricks-workspace/relocation.csv"

csv_path = "/Users/sawan.darekar/Downloads/MI_state_revert_relocations_scope.csv"

relocation_df = pd.read_csv(csv_path)

print(f"Relocation CSV row count: {len(relocation_df)}")
display(relocation_df)
```

### Generate relocation-scope.csv — 4 APA tag rows per Product_orbis_id

```python
import numpy as np

APA_TAGS = [
    "metadata:apa:accuracy",
    "metadata:apa:previous_location",
    "metadata:apa:source",
    "metadata:apa:updated",
]

orbis_ids = relocation_df["Product_orbis_id"].unique()

scope_df = pd.DataFrame({
    "Product_orbis_id": np.repeat(orbis_ids, len(APA_TAGS)),
    "Product_Address_component": APA_TAGS * len(orbis_ids),
    "Product_Language": "",
    "Source_Feature_Id": "",
    "Target_Value": "",
    "Orbis_X": 0,
    "Orbis_Y": 0,
})

output_path = "/Users/sawan.darekar/Desktop/tomtom_git_workspace/databricks-workspace/relocation-updated-all-scope.csv"
scope_df.to_csv(output_path, index=False)

print(f"Input  rows : {len(relocation_df)}")
print(f"Output rows : {len(scope_df)}  ({len(orbis_ids)} unique IDs × {len(APA_TAGS)} tags)")
display(scope_df.head(8))
```
