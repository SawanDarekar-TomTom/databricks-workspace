```python
%scala
import org.apache.spark.sql.functions.{col, count}

val msftBfp = spark.read.table("apa_prod.msft_bfp")
val osmBfp  = spark.read.table("apa_prod.osm_bfp")

println("=== apa_prod.msft_bfp — count by country_iso ===")
msftBfp.groupBy("country_iso")
  .agg(count("*").alias("cnt"))
  .orderBy(col("cnt").desc)
  .show(300, truncate = false)

println("=== apa_prod.osm_bfp — count by country_iso ===")
osmBfp.groupBy("country_iso")
  .agg(count("*").alias("cnt"))
  .orderBy(col("cnt").desc)
  .show(300, truncate = false)
```

```python
%scala
import org.apache.spark.sql.functions.{col, count, coalesce, lit}

val msftCounts = spark.read.table("apa_prod.msft_bfp")
  .groupBy("country_iso").agg(count("*").alias("msft_count"))

val osmCounts = spark.read.table("apa_prod.osm_bfp")
  .groupBy("country_iso").agg(count("*").alias("osm_count"))

msftCounts.join(osmCounts, Seq("country_iso"), "full_outer")
  .withColumn("msft_count", coalesce(col("msft_count"), lit(0)))
  .withColumn("osm_count",  coalesce(col("osm_count"),  lit(0)))
  .orderBy(col("country_iso"))
  .show(300, truncate = false)
```
